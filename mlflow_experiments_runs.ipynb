```python
from mlflow.tracking import MlflowClient

client = MlflowClient()
client
```

```python
client.get_experiment_by_name("W1D2")
```

```python
# get the list of experiments
experiments = client.search_experiments()
```

```python
experiments
```

```python
exp_id = experiments[0].experiment_id
exp_name = experiments[0].name
print(exp_id,exp_name)
```

```python
import mlflow
```

```python
# get runs for a specific experiment
runs = mlflow.search_runs(experiment_names=[exp_name])
runs
```

```python
# getting info for a specific run
client.get_run(runs[0:1].run_id.values[0])
```

```python
mlflow.set_experiment(exp_name)
runid = runs[0:1].run_id.values[0]
with mlflow.start_run(run_id = runid):
    mlflow.log_metric("test_accuracy",0.8)
    mlflow.log_metric("test_recall",0.7)
```
